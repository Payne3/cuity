```python
%matplotlib inline
%matplotlib notebook
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
crime_data = "Crimes_-_2018.csv"
housing_data = "Affordable_Rental.csv"

crime_df = pd.read_csv(crime_data)

housing_df = pd.read_csv(housing_data)


```

```python
# drop Nan values and uneccessary columns from data

crime_df = crime_df.dropna()
crime_df
crime_df = crime_df.drop(["Census Tracts", "Wards", "Location", "Community Areas","Historical Wards 2003-2015", "Boundaries - ZIP Codes", "Updated On", "Police Districts", "Police Beats", "FBI Code", "X Coordinate", "Y Coordinate", "District"] ,axis=1 )
crime_df
```

```python
crime_df = crime_df.rename(columns= {"Case Number":"Case_Number", "Date":"date_", "Block":"block_", "Primary Type":"Primary_Type", "Location Description": "Location_Description", "Community Area": "Community_Area", "Year": "Year_", })
```

```python
crime_df["date_"] = pd.to_datetime(crime_df["date_"])
print(crime_df.dtypes)
```

```python
crime_df["date_"] = pd.to_datetime(crime_df["date_"])
print(crime_df.dtypes)
```
